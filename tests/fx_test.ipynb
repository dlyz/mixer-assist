```python
from xair_client.client import XAirConnection
from xair_client.nodes_base import MixerNode, MixerPropertyNode
from xair_client.text_tree_service import MixerTextTreeService
from xair_client.nodes.mixer import Mixer
from xair_client.nodes.effects.all import FxType
```

```python
ip = os.environ["MIXER_IP"]
client = XAirConnection(ip=ip, timeout=1.0).connect()

print(f"Connected to {ip}, mixer model: {client.mixer_model.id}")
```

```python
def print_fx_batch(fx: int):
    mixer = Mixer(client)
    text_tree = MixerTextTreeService(mixer)

    fx_slot = 1
    type = list(FxType)[fx]
    mixer.fx[fx_slot].effect_type = type

    print(type.name)
    print(text_tree.expand_node(f"/fx/{fx_slot}/effect_params", verbose=True))
    print()
```

```python
print_fx_batch(60)
```
